```python
from jetbot import Robot
import time
import os #used for removing the command file once it is read from

import os.path
from os import path #for checking if there's a command file

robot = Robot()
import socket # for connecting to server
```

```python
desk = '192.168.1.102'
lap = '127.0.0.1'
lap2 = "192.168.1.30"
host = lap2 #Set this to be your IP address, found in ipconfig or socket.gethostbyname(socket.gethostname())
while True:
    #Connecting
    s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    s.connect((host, 1234)) #IPv4 Address OF THE SERVER found using ipconfig in terminal
    #Receiving Data
    msg = s.recv(1024)
    NewCommand = msg.decode("utf-8")
    power = 0.5
    duration = 0.1
    print(NewCommand)
    
    if NewCommand == 'left':   #Do what the file says, left right or forward
        print("Turning Left")
        robot.left(speed = power)
        time.sleep(duration)
        robot.stop()
    elif NewCommand == 'right':
        print("Turning Right")
        robot.right(speed = power)
        time.sleep(duration)
        robot.stop()    
    elif NewCommand == 'forward':
        print("Going forward")
        robot.forward(speed = power)
        time.sleep(duration)
        robot.stop()  
    time.sleep(1)
```
